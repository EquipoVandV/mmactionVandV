# Prepare Dataset

```python
######### Convert Videos to pkls with pose detections#######

import os 
from IPython.display import clear_output


vid_dir='C:/Users/Hidrolatina/Downloads/Videos_test/dataset2/'  #Folder with videos with names ending in Axx, with xx as the action number
name_vid=[vid_dir+item for item in os.listdir(vid_dir)]
pkl_dir='C:/Users/Hidrolatina/Downloads/Videos_test/dataset89a/' #Folder to place individual pkls of each video pose detection
if not os.path.isdir(pkl_dir):
    os.mkdir(pkl_dir)
    print("Pkl folder does not exist but was created")
name_pkl= [pkl_dir+item.split('.')[0]+'.pkl' for item in os.listdir(vid_dir)]

for i in range(len(name_vid)):
    os.environ['vid'] = name_vid[i]
    os.environ['pkl'] = name_pkl[i]
    ##Pose extraction Script directory
    %run ../tools/data/skeleton/Pose_Extraction_MaxArea.py
    clear_output(wait=False)
    print('\nVideo {}/{} has been converted'.format(i+1, len(name_vid)))

print('\n Conversion Completed')


 
```

```python
####Split and append pkl dataset list in Train and Validation
from mmcv import load, dump
pkl_dir='C:/Users/Hidrolatina/Downloads/Videos_test/3HL_pkl//' #Folder with individuals pkls of pose detections
pkls=[pkl_dir+item for item in os.listdir(pkl_dir)]
pt=int(.8*len(pkls)) #80% Train -  20% Validation
pkl_train=pkls[:pt]
pkl_val=pkls[pt:len(pkls)]
# print(train_list)
# print(val_list)
train_list=val_list=[]
for name in pkl_train:
    train_list.append(load(name))
for name in pkl_val:
    val_list.append(load(name))
dump(train_list, 'C:/Hidrolatina/mmaction2/Train/data/PoseC3D/3HL_train.pkl')    #Train list dataset
dump(val_list, 'C:/Hidrolatina/mmaction2/Train/data/PoseC3D/3HL_val.pkl')   #Validation list dataset
```
